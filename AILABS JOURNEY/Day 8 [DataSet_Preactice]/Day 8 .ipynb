# concat()

```python
indian_weather = pd.DataFrame({"city":["mumbai","delhi","Bangalore"],"temperature":(32,45,30),"humidity":[50,60,70]})
print(indian_weather)

us_weather = pd.DataFrame({"city":["Salt Lake","Los Angeles","New York"],"temperature":(17,35,60),"humidity":[10,30,28]})
print(us_weather)
```

```python
var  = pd.concat([indian_weather, us_weather], keys=["India", 'US'])
var
```

```python
var.loc['India',0]
```

```python
var.loc["US"].loc[0] #to get the inside values
```

```python
var.loc["India"].loc[2] #to get the inside values
```
