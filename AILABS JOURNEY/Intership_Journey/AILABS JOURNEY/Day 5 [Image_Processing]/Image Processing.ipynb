# Image Processing

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
ai_man = io.imread("ai.jpeg")
print(ai_man)
```

```python
print(ai_man.shape)
plt.imshow(ai_man)
plt.show()
```

Crop the image

```python
cropped = ai_man[100:450:,180:470,:]
plt.imshow(cropped)
plt.show()
```

```python
io.imsave('ai_man_cropped.jpg',cropped)
```

Flip the image vertically

```python
flip = ai_man[::-1,:,:]
plt.imshow(flip)
plt.show()
```

Flip the image horizontally

```python
flip = ai_man[:,::-1,:]
plt.imshow(flip)
plt.show()
```
