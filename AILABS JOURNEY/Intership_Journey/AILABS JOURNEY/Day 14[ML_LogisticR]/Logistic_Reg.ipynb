# Logistic Regression

```python
import pandas as pd
import numpy as np
```

```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
datas = datasets.load_breast_cancer()
```

```python
x = datas.data
```

```python
y = datas.target
```

```python
x1 = pd.DataFrame(x)
```

```python
f = datas.feature_names
```

```python
len(f)
```

```python
x1.columns = f
```

```python
x1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x1,y,test_size =  0.3, random_state=1)
```

```python
X_test
```

```python
X_train.shape
```

```python
X_train.index
```

```python
obj = LogisticRegression(C = 2.1)
obj.fit(X_train,y_train)
```

```python
y_test
```

```python
y_test = obj.predict(X_test)
```

```python
obj.score(X_train,y_train)
```
