```python
import pandas as pd
from sklearn import datasets
datas = datasets.load_iris()
datas.data
x = datas.data      #x is feature set, (which describe the data)
y = datas.target
df = pd.DataFrame(x)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import model_selection
X_train,X_test,y_train,y_test = model_selection.train_test_split(x,y,test_size= 0.2, random_state=46) #test_size = percentage of spliting.  Random_state is used to set the seed for the random generator so that we can ensure that the results that we get can be reproduced. Because of the nature of splitting the data in train and test is randomised you would get different data assigned to the train and test data unless you can control for the random factor.
```

```python
dtc = DecisionTreeClassifier(min_samples_split=10,min_samples_leaf=5)   #to make model accurate.
# dtc = DecisionTreeClassifier() 
```

```python
dtc.fit(X_train,y_train)
```

```python
dtc.predict(X_train)
```

```python
dtc.predict(X_test)   #here no need to give y_test because this is the o/p. 
```

```python
dtc.score(X_train, y_train)
```

```python
dtc.score(X_test,y_test)
```
