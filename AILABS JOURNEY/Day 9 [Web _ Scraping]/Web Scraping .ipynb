```python
# Day 9  || Web Scraping

                                                                                                                06-07-2023

import requests #works on HTTP protocol 

requests.get('https://books.toscrape.com/')  #<Response [200]> means it OK. #it return a function (because we don't write print fuction)

requests.get('https://books.toscrape.com/12345') #this web site is not exit , so that it occur error.

obj = requests.get('https://books.toscrape.com/')
obj

print(obj.content)

from bs4 import BeautifulSoup    #BeautifulSoup use to convert from string form text to html form.

bs = BeautifulSoup(obj.content,'html.parser')  #use of html.parser is use to read as html document 
bs

print(bs.prettify())  #prettify use to look nice exact fornat.

bs.find('ol')

bs.find('pclass')

ol = bs.find('ol')
ol

pod = ol.find('article',class_ = 'product_pod')
pod

ol.find_all()

img = pod.find('img')
img

img.attrs

img.attrs['alt']

```

```python
title_list = []
pricing_list = []
for i in pod:
    img = i.find("img")
    pricing = i.find("p", class_ = "price_color")
    pricing_list.append(float(pricing.contents[0][1:]))
    title_list.append(img.attrs["alt"])
    print(title_list)
    print(pricing_list)
```

```python
import pandas as pd
```

```python
result_list = []
for i in pod:
    img = i.find("img")
    pricing = i.find("p", class_ = "price_color")
    rating = i.find("p")
    var = (img.attrs["alt"],float(pricing.contents[0][1:]), rating.attrs['class'][1])
    result_list.append(var)
print(result_list)
pd.DataFrame(result_list)
```
